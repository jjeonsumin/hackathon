```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import matplotlib.font_manager as fm


plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False 
warnings.filterwarnings('ignore')

# 파일 로드
data = pd.read_csv("C:/apps/hct/datasets/하이닉스 power.csv", encoding="cp949")

# Scaler 인스턴스 생성
# min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
robust_scaler = RobustScaler()

# # Min-Max Scaling 적용할 컬럼
# min_max_columns = [
#     'SMA_5', 'SMA_20', 'SMA_60', 'SMA_120', 'SMA_240',
#     'EMA_5', 'EMA_20', 'EMA_60', 'EMA_120', 'EMA_240', 'SMA_5_minus_SMA_20',
#     'SMA_5_minus_SMA_60', 'SMA_5_minus_SMA_120', 'SMA_5_minus_SMA_240',
#     'SMA_20_minus_SMA_60', 'SMA_20_minus_SMA_120', 'SMA_20_minus_SMA_240',
#     'SMA_60_minus_SMA_120', 'SMA_60_minus_SMA_240', 'SMA_120_minus_SMA_240',
#     'EMA_5_minus_EMA_20', 'EMA_5_minus_EMA_60', 'EMA_5_minus_EMA_120', 'EMA_5_minus_EMA_240',
#     'EMA_20_minus_EMA_60', 'EMA_20_minus_EMA_120', 'EMA_20_minus_EMA_240',
#     'EMA_60_minus_EMA_120', 'EMA_60_minus_EMA_240', 'EMA_120_minus_EMA_240',
#     '종가_minus_SMA_5', '종가_minus_SMA_20', '종가_minus_SMA_60', '종가_minus_SMA_120', '종가_minus_SMA_240',
#     '종가_minus_EMA_5', '종가_minus_EMA_20', '종가_minus_EMA_60', '종가_minus_EMA_120', '종가_minus_EMA_240'
# ]

# Standard Scaling 적용할 컬럼
standard_columns = [
    'MACD', 'MACD_Signal', 'RSI', 'RSI_Signal', 
    '시가_등락률', '고가_등락률', '저가_등락률', '종가_등락률', '거래량_등락률',
    'MACD_등락률', 'MACD_Signal_등락률', 'RSI_등락률', 'RSI_Signal_등락률',
    'SMA_5_등락률', 'SMA_20_등락률', 'SMA_60_등락률', 'SMA_120_등락률', 'SMA_240_등락률',
    'EMA_5_등락률', 'EMA_20_등락률', 'EMA_60_등락률', 'EMA_120_등락률', 'EMA_240_등락률', 'SMA_5', 'SMA_20', 'SMA_60', 'SMA_120', 'SMA_240',
    'EMA_5', 'EMA_20', 'EMA_60', 'EMA_120', 'EMA_240', 'SMA_5_minus_SMA_20',
    'SMA_5_minus_SMA_60', 'SMA_5_minus_SMA_120', 'SMA_5_minus_SMA_240',
    'SMA_20_minus_SMA_60', 'SMA_20_minus_SMA_120', 'SMA_20_minus_SMA_240',
    'SMA_60_minus_SMA_120', 'SMA_60_minus_SMA_240', 'SMA_120_minus_SMA_240',
    'EMA_5_minus_EMA_20', 'EMA_5_minus_EMA_60', 'EMA_5_minus_EMA_120', 'EMA_5_minus_EMA_240',
    'EMA_20_minus_EMA_60', 'EMA_20_minus_EMA_120', 'EMA_20_minus_EMA_240',
    'EMA_60_minus_EMA_120', 'EMA_60_minus_EMA_240', 'EMA_120_minus_EMA_240',
    '종가_minus_SMA_5', '종가_minus_SMA_20', '종가_minus_SMA_60', '종가_minus_SMA_120', '종가_minus_SMA_240',
    '종가_minus_EMA_5', '종가_minus_EMA_20', '종가_minus_EMA_60', '종가_minus_EMA_120', '종가_minus_EMA_240', '[일]국가지자체_수량'
]

# Robust Scaling 적용할 컬럼
robust_columns = [
    '시가', '고가', '저가', '종가','거래량', '[일]사모펀드_수량', '[일]기타법인_수량', '[일]연기금 등_수량', '[일]기타금융_수량', '[일]은행_수량',
    '[일]투신_수량', '[일]보험_수량', '[일]금융투자_수량', '[일]기관_수량', 
    '[일]외국인_수량', '[일]개인_수량', '[일]프로그램순매수(종목)'
]
for col in standard_columns + robust_columns:
    if data[col].dtype == object:
        data[col] = data[col].str.replace(',', '').astype(float)

# Min-Max Scaling 적용
# data[min_max_columns] = min_max_scaler.fit_transform(data[min_max_columns])

# Standard Scaling 적용
data[standard_columns] = standard_scaler.fit_transform(data[standard_columns])

# Robust Scaling 적용
data[robust_columns] = robust_scaler.fit_transform(data[robust_columns])

data.to_csv("C:/apps/hct/datasets/하이닉스 scaling2.csv", encoding="cp949")
# save_path = 'C:/apps/hct/datasets/하이닉스 power.csv'
# data.to_csv(save_path, index=False, encoding='utf-8-sig')

# 스케일링 결과 확인
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False 
warnings.filterwarnings('ignore')

rows = 18
cols = 5
fig, axs = plt.subplots(rows, cols, figsize=(20, 44)) # 전체 그래프의 크기 조절

# 각 컬럼에 대해 산점도를 그림
col_index = 1  # 'a' 컬럼부터 시작
for i in range(rows):
    for j in range(cols):
        if col_index < len(data.columns):
            sns.scatterplot(ax=axs[i, j], x=data.index, y=data[data.columns[col_index]])
            axs[i, j].set_title(data.columns[col_index])
            axs[i, j].set_xlabel('Index')
            axs[i, j].set_ylabel('')
        else:
            axs[i, j].axis('off')  # 사용하지 않는 서브플롯은 숨김
        col_index += 1

plt.tight_layout()  # 서브플롯 간 간격 조정
plt.show()  # 그래프 표시
```
